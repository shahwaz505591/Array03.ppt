    closestSum = float('inf')
    
    for i in range(n - 2):
        left = i + 1
        right = n - 1

            currentSum = nums[i] + nums[left] + nums[right]

    def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return left
